# **Rock vs Mine Prediction**

# Using Linear Regression

```python
#importing dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
```

```python
#data collection and processing 

#loading dataset
sonarData = pd.read_csv('/kaggle/input/sonar-data/SonarData.csv', header=None)
```

```python
sonarData.head()
```

```python
#number of rows and coloumns
sonarData.shape
```

```python
sonarData.describe()
```

```python
sonarData[60].value_counts()
```

```python
sonarData.groupby(60).mean()
```

```python
#separating data and labels

X = sonarData.drop(columns=60, axis=1)
Y = sonarData[60]
print(X)
print(Y)
```

```python
#Training and test data 

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=3)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
#model training

model.fit(x_train, y_train)
```

```python
x_train_pred = model.predict(x_train)
train_data_accuracy = accuracy_score(x_train_pred, y_train)
```

```python
print(train_data_accuracy)
```

```python
x_test_pred = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred, y_test)
print(test_data_accuracy)
```

# Using Random Forest

```python
rf_model = RandomForestClassifier(n_estimators=500, random_state=42)
```

```python
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)

acc = accuracy_score(y_test, y_pred)
print(f"Random Forest Model Accuracy: {acc * 100:.2f}%")
```

```python
# Predict on the training data
y_trn_pred = rf_model.predict(x_train)

# Calculate the training accuracy
train_accuracy = accuracy_score(y_train, y_trn_pred)
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
```

# Using CatBoost

```python
# Initialize the CatBoost Classifier
catboost_model = CatBoostClassifier(iterations=500, learning_rate=0.1, depth=6, verbose=0)

# Train the model
catboost_model.fit(x_train, y_train)
```

```python
# Predict on the test data
y_test_pred = catboost_model.predict(x_test)

# Calculate the test accuracy
test_accuracy = accuracy_score(y_test, y_test_pred)
print(f"CatBoost Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
# Predict on the training data
y_train_pred = catboost_model.predict(x_train)

# Calculate the training accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"CatBoost Training Accuracy: {train_accuracy * 100:.2f}%")
```
